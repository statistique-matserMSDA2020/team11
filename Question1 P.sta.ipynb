##### *Loi Binomiale*#####

```python
import numpy as np
import scipy.stats as sps
import matplotlib.pyplot as plt
n, p, N = 30, 0.2, 10000
B = np.random.binomial(n, p, N)
f = sps.binom.pmf(np.arange(n+1), n, p)
plt.hist(B,bins=n+1,normed=1,range=(0.5,n+.5),label="loi empirique")
plt.stem(np.arange(n+1),f,"r",label="loi theorique")
plt.legend()
plt.show()
```

#### *Loi Normale*#####

```python
import numpy as np
import matplotlib.pyplot as plt
mu, sigma = 3, .4 
s = np.random.normal(mu, sigma, 10000)
count, bins, ignored = plt.hist(s, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (bins - mu)**2 / (2 * sigma**2)),
         linewidth=2, color='r', label ="loi normale")
plt.legend()
plt.show()
```

##### *Loi Gamma*#####

```python
import scipy.stats as stats 
from matplotlib import pyplot as plt
x = np.linspace (0, 100,200) 
y1 = stats.gamma.pdf(x, a=29, loc=3)
plt.plot(x, y1, "y-", label=("Loi gamma")) 


plt.ylim([0,0.08])
plt.xlim([0,150])
plt.legend()
plt.show()
```
